```python
#question 3 Numpy is not well suited to loading/manipulating strings. We therefore have to find a different
#way to read the category string values from categories.csv than using a numpy function. Import the
#csv module and extract the category value strings and store it in a list.
#Hint The python function open and the csv function csv.reader are useful. The function
#csv.reader returns an object that is iterable, that is we can use the for loop to explore its contents
#one by one

import csv
list_cat =[]
    
with open ("categories.csv") as csv_file:
    cat = csv.reader(csv_file, delimiter =',')
    for row in cat:
        list_cat.append(row)

```
